```python
# concert  runtime_pred['segment'] to integer
runtime_pred['segment'] = runtime_pred['segment'].astype(int)
runtime_pred['trip_id'] = runtime_pred['trip_id'].astype(int)
```

```python

bus_stops = ['BT01', '101', '102', '103', '104', '105', '106', '107', '108', '109', '110', '111', '112', '113', '114', 'BT02']

# Sort the dataframe by 'trip_id' and 'segment' to ensure correct order
runtime_pred.sort_values(by=['trip_id', 'segment'], inplace=True)

# Calculate 'from_bus_stop' and 'to_bus_stop' based on 'segment'
runtime_pred['from_bus_stop'] = runtime_pred['segment'].apply(lambda x: bus_stops[x - 1] if x != 1 else bus_stops[0])
runtime_pred['to_bus_stop'] = runtime_pred['segment'].apply(lambda x: bus_stops[x] if x != len(bus_stops) - 1 else bus_stops[-1])

runtime_pred

```

```python
# rename the bus_stop column in dwelltime_pred to from_bus_stop
dwelltime_pred.rename(columns={'bus_stop': 'from_bus_stop'}, inplace=True)  
```

```python
import pandas as pd

# Assuming you have two dataframes: dwelltime_pred and runtime_pred
# dwelltime_pred has columns: 'trip_id', 'bus_stop', 'predicted_dwell_time'
# runtime_pred has columns: 'trip_id', 'segment', 'predicted_run_time'

# Convert 'from_bus_stop' column to string data type
dwelltime_pred['from_bus_stop'] = dwelltime_pred['from_bus_stop'].astype(str)
runtime_pred['from_bus_stop'] = runtime_pred['from_bus_stop'].astype(str)

# Merge the two dataframes based on 'trip_id' and 'from_bus_stop' columns
final_df = pd.merge(dwelltime_pred, runtime_pred, on=['trip_id', 'from_bus_stop'], how='right')

```

```python
final_df
```

```python
# fill nan in predicted_dwell_time with 0
final_df['predicted_dwell_time'].fillna(0, inplace=True)
```

```python
final_df['total_time'] = final_df['predicted_dwell_time'] + final_df['predicted_run_time']
```

```python
total_time_df=final_df[['trip_id','deviceid_x','deviceid_y','from_bus_stop','to_bus_stop','predicted_dwell_time','predicted_run_time','total_time']]
```

```python
# total_time_df drop nan
total_time_df.dropna(inplace=True)
```

```python
bustop_locations_kandy_digana
```

```python
total_time_df = pd.merge(total_time_df, bustop_locations_kandy_digana, how='left', left_on='from_bus_stop', right_on='stop_id')
```

```python
total_time_df
```

```python
# save to csv
total_time_df.to_csv('../DataOut/total_time.csv', index=False)
```

```python
# total_time is in seconds , convert in to minutes in panas date time
total_time_df['total_time'] = pd.to_datetime(total_time_df['total_time'], unit='s').dt.strftime("%H:%M:%S")
```
